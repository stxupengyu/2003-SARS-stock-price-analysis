```python
import tushare as ts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
sz=ts.get_k_data('sz',start='2003-01-01', end='2003-12-30',autype='qfq')
sh=ts.get_k_data('sh',start='2003-01-01', end='2003-12-30',autype='qfq')
```

```python
date=sh['date']
sz=sz['close']
sh=sh['close']
```

```python
date[date=='2003-02-11']
```

```python
a1=date[date=='2003-02-11']#2月11日，广东省主要媒体报道了部分地区先后发生非典型肺炎病例的情况
a2=date[date=='2003-03-25']#医务人员死亡
a3=date[date=='2003-04-15']#山西 广州 香港 被列为疫区
a4=date[date=='2003-04-21']#瞒报被处理 高校停课
a5=date[date=='2003-05-12']#5月2日到5月8日，平均每天6.3人。北京非典病例呈大幅下降趋势
a6=date[date=='2003-05-19']#5月17日，北京大学人民医院解除隔离
a7=date[date=='2003-05-23']#5月23日，北京市747名密切接触者全部解除隔离，非典传播链完全切断
a8=date[date=='2003-06-02']#2003年6月1日，卫生部宣布北京市防治非典型肺炎指挥部撤销
a9=date[date=='2003-06-16']#6月14日，WHO解除对河北、内蒙古、山西、天津的旅游警告
```
